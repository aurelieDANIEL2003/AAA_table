```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re


url = "https://www.yelp.com/biz/sunday-in-soho-paris?adjust_creative=NnnrNySybEItfMzq-c4qFw&utm_campaign=yelp_api_v3&utm_medium=api_v3_business_search&utm_source=NnnrNySybEItfMzq-c4qFw"
navigator = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1)'
html = requests.get(url, headers={'User-Agent': navigator})
soup = BeautifulSoup(html.text, 'html.parser')

liste_avis = []

for balise_parent in soup.find_all('span', {"class": "y-css-r8orer"}):
  for element in balise_parent.find_all('a', {"class": "y-css-1x1e1r2"}):
    liste_avis.append(element.get_text().strip())

liste_avis
```

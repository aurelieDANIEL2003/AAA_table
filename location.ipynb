```python
import pandas as pd
```

```python
ville= pd.read_csv('csv/cities.csv')
```

```python
ville1 = ville.drop(columns=['slug','insee_code', 'gps_lat','gps_lng','zip_code'])
ville1
```

```python
departement = pd.read_csv('csv/departments.csv')
departement = departement.drop(columns=['slug','region_code'])
departement
```

```python
df_loc = pd.merge(departement, ville1, left_on='code', right_on='department_code', how='inner')
df_loc = df_loc.drop(columns=['id_x','id_y', 'code'])
df_loc.rename(columns={"name_x": "nom_departement", "name_y": "nom_ville"}, inplace=True)
df_loc
```

```python
df_loc = df_loc.drop_duplicates()
```

```python
import requests
import re

def extraire_avis(data):
    results = []

    # Parcourir les éléments dans data['items']
    for item in data.get('items', []):
        title = item.get('title', 'Titre non disponible')
        link = item.get('link', 'Lien non disponible')
        snippet = item.get('htmlSnippet', '')

        # Chercher le nombre d'avis dans le snippet (exemple : "(386 avis)")
        avis_match = re.search(r"(?:voté|rated)\s*([\d.,]+)", snippet)
        nb_avis = avis_match.group(1) if avis_match else "Non disponible"

        if nb_avis != "Non disponible":
            # Ajouter le résultat
            
            results.append({
                'title': title,
                'link': link,
                'nb_avis': nb_avis
                         })
        else:
            print("Pas d'avis")

    return results


# Exemple d'utilisation
name = "Le Siècle d'Or"
location = "Montélimar"
APIcleG = "AIzaSyDIs6LZdQTHiXQi3ETWJ_ZU6JM4ZwBZAwc"
CX = "232dc5f95baca4e2c"
query = f"{name} {location}"
urlG = f"https://www.googleapis.com/customsearch/v1?key={APIcleG}&cx={CX}&q={query}"

response = requests.get(urlG)
if response.status_code == 200:
    data = response.json()
    avis_results = extraire_avis(data)

    # Afficher les résultats
    for avis in avis_results:
        print(f"Titre : {avis['title']}")
        print(f"Lien : {avis['link']}")
        print(f"Note : {avis['nb_avis']}")
        print("---")
else:
    print(f"Erreur HTTP : {response.status_code}")
```
